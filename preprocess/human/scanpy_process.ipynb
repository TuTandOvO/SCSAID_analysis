```python
def merge_GSE_dicts(*gse_dicts):
    all_adatas = []
    for gse_dict in gse_dicts:
        all_adatas.extend(list(gse_dict.values()))
    
    merged = ad.concat(all_adatas, join="outer", index_unique=None)
    return merged

adata_merged = merge_GSE_dicts(GSE249793, GSE155850, GSE175990, GSE179565, GSE130973,
                               GSE242477, GSE249622, GSE203034, GSE193304, GSE195674,
                               GSE237754, GSE245859, GSE248121, GSE213849, GSE233137,
                               HRA000166, GSE132802, GSE203262, GSE158924, GSE181297,
                               GSE275846, GSE212447, GSE173252)
adata_merged.var_names_make_unique()
adata_merged.obs_names_make_unique()
```

```python
# 统计每个 batch 的细胞数
batch_sizes = adata_merged.obs["batch"].value_counts()
print(batch_sizes.sort_values())
min_cells = 350
keep_batches = batch_sizes.index[batch_sizes >= min_cells]
adata_merged = adata_merged[adata_merged.obs['batch'].isin(keep_batches)].copy()
adata_merged.obs_names_make_unique()
# 按照细胞数升序排序
print("after filter")
batch_sizes = adata_merged.obs["batch"].value_counts()
print(batch_sizes.sort_values())
```
